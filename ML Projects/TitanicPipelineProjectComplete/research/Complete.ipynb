```python
y = df['Survived']
x = df.drop(columns =['Survived'], axis = 1)
```

```python
# prompt: Using dataframe x: scalling of data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```
